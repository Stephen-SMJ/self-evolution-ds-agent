# Utils

```python
import requests

def get_categorymembers(title, cmcontinue=None):
    base_url = "https://en.wikipedia.org/w/api.php"

    params = {
        "action": "query",
        "list": "categorymembers",
        "cmtitle": title,
        "cmlimit": "500",
        "format": "json",
    }

    if cmcontinue:
        params["cmcontinue"] = cmcontinue

    response = requests.get(base_url, params=params)

    return response.json()

def get_all_categorymembers(title, depth=1):
    page_urls = []

    # Check the depth
    if depth < 0:
        return page_urls

    result = get_categorymembers(title)

    while True:
        for member in result['query']['categorymembers']:
            page_title = member['title'].replace(' ', '_')
            page_url = f"https://en.wikipedia.org/wiki/{page_title}"
            
            # Check if category member is a page or a subcategory
            if 'Category:' in member['title']:
                # If it's a subcategory, fetch its members recursively
                page_urls.extend(get_all_categorymembers(member['title'], depth=depth-1))
            else:
                # If it's a page, add it to the list
                page_urls.append(page_url)

        if 'continue' in result:
            cmcontinue = result['continue']['cmcontinue']
            result = get_categorymembers(title, cmcontinue)
        else:
            break

    return page_urls
```

```python
URL_BANK = set()
```

# Physics

```python
category2depth = {
    "Category:Concepts_in_physics": 3,
    "Category:Subfields of physics": 2,
    "Category:Physical_systems": 2,
    "Category:Physical modeling": 1,
}
```

```python
%%time
url_list = []

for cat, depth in category2depth.items():
    print(f"Getting URLs for {cat} with depth {depth}")
    urls = get_all_categorymembers(cat, depth=depth)
    print(f"# urls extracted = {len(urls)}")
    url_list.append(urls)
```

```python
physics_urls = deepcopy(list(chain(*url_list)))
physics_urls = list(set(physics_urls))

len(physics_urls)
```

```python
URL_BANK.update(physics_urls)
print(f"size of URL bank {len(URL_BANK)}")
```

```python
random.sample(physics_urls, 25)
```
