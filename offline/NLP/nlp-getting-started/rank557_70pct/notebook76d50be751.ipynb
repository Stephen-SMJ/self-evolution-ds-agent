```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

class DisasterTweetClassifier:
    """
    An Object-Oriented NLP Classifier to detect disaster tweets 
    using TF-IDF Vectorization and Naive Bayes.
    """
    def __init__(self):
        # Initialize the TF-IDF Vectorizer to convert text to mathematical vectors
        self.vectorizer = TfidfVectorizer(stop_words='english')
        # Initialize the Naive Bayes model for classification
        self.model = MultinomialNB()

    def train_model(self, X_train_text, y_train_labels):
        print("1. Converting text to vectors and training the model...")
        # Fit the vectorizer to learn the vocabulary and transform the training text
        X_train_vectors = self.vectorizer.fit_transform(X_train_text)
        # Train the Naive Bayes model
        self.model.fit(X_train_vectors, y_train_labels)
        print("   -> Training complete! Model is ready.")

    def make_predictions(self, X_test_text):
        print("2. Predicting on test data...")
        # Transform the test text using the ALREADY FITTED vectorizer
        X_test_vectors = self.vectorizer.transform(X_test_text)
        # Return the predicted labels (0 or 1)
        return self.model.predict(X_test_vectors)


# ==========================================
# EXECUTION WORKFLOW
# ==========================================

# 1. Load the data (Kaggle automatically stores it in the input folder)
train_path = '/kaggle/input/competitions/nlp-getting-started/train.csv'
test_path = '/kaggle/input/competitions/nlp-getting-started/test.csv'

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

# 2. Instantiate the OOP class
tweet_classifier = DisasterTweetClassifier()

# 3. Train the model using the 'text' and 'target' columns
tweet_classifier.train_model(train_df['text'], train_df['target'])

# 4. Generate predictions for the test dataset
predictions = tweet_classifier.make_predictions(test_df['text'])

# 5. Create the submission DataFrame exactly as Kaggle requires
submission_df = pd.DataFrame({
    'id': test_df['id'],
    'target': predictions
})

# 6. Save the submission file
submission_df.to_csv('submission.csv', index=False)
print("3. submission.csv created successfully! Ready for upload to the leaderboard.")
```
