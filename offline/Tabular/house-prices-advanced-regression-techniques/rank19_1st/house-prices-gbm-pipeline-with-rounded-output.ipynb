```python
import joblib
import json
import numpy as np
import pandas as pd
from IPython.display import display

def load_model(path):
    return joblib.load(path)

def load_data(path):
    return pd.read_csv(path)

def round_to_100(values):
    return (np.round(values / 100) * 100).astype(int)

def create_submission(model, test_path, sample_path, output_path):
    test_df = load_data(test_path)
    preds = model.predict(test_df)
    preds = np.expm1(preds)
    preds = round_to_100(preds)
    submission = load_data(sample_path)
    submission["SalePrice"] = preds
    submission.to_csv(output_path, index=False)
    return submission

model = load_model("/kaggle/input/models/jek1wantaufik/buddy/scikitlearn/house/1/output/model.joblib")

submission = create_submission(
    model,
    "/kaggle/input/competitions/house-prices-advanced-regression-techniques/test.csv",
    "/kaggle/input/competitions/house-prices-advanced-regression-techniques/sample_submission.csv",
    "/kaggle/working/submission.csv"
)

display(submission.head())
```
