## 1. 🧭 Problem Setup

<div style="background:#FFFCF8; border:1px solid #E2D8C7; border-left:6px solid #2F6B4F; border-radius:14px; padding:13px 15px; margin:10px 0 16px 0; color:#233127; line-height:1.58; box-shadow:0 8px 18px rgba(35,49,39,0.04);">Spaceship Titanic is a <b>binary classification</b> problem. For each passenger, the model predicts whether <span style="color:#2F6B4F; font-weight:900; border-bottom:2px solid #CFE0D0;">Transported</span> is <b>True</b> or <b>False</b>. The public score is accuracy.</div>

| Concept | Meaning in this competition | Why it matters |
|---|---|---|
| 🧍 Row | One passenger | Each row receives one prediction. |
| 🧩 Feature | A column used as input | Features describe identity, travel plan, cabin, demographics, and spending. |
| 🎯 Target | <span style="color:#2F6B4F; font-weight:900;">Transported</span> | This is the label the model learns from in `train.csv`. |
| 🧪 Test set | Rows without target labels | Kaggle evaluates predictions for these rows. |
| ✅ Accuracy | Fraction of correct labels | This competition rewards exact `True` or `False` predictions. |


```python
# =================================================================================
# LIBRARIES, GLOBALS, AND DISPLAY HELPERS
# =================================================================================
from pathlib import Path
from html import escape
import inspect
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.ticker import MaxNLocator
import seaborn as sns

from IPython.display import HTML, display
from sklearn.base import clone
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import (
    ExtraTreesClassifier,
    HistGradientBoostingClassifier,
    RandomForestClassifier,
    VotingClassifier,
)
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.inspection import permutation_importance
from sklearn.metrics import accuracy_score, confusion_matrix, log_loss, roc_auc_score
from sklearn.model_selection import StratifiedGroupKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 160)
pd.set_option('display.max_rows', 140)
pd.set_option('display.width', 220)

COLORS = {
    'primary': '#2F6B4F',
    'accent': '#C7772C',
    'risk': '#B5473C',
    'signal': '#3D6F8A',
    'neutral': '#7A8571',
    'dark': '#233127',
    'surface': '#F7F3EA',
    'surface_alt': '#EEF4ED',
    'panel': '#FFFCF8',
    'table_alt': '#F6F3EC',
    'border': '#D9D0BF',
}

sns.set_theme(
    style='whitegrid',
    context='paper',
    palette=[COLORS['primary'], COLORS['accent'], COLORS['signal'], COLORS['risk']],
    rc={
        'figure.facecolor': 'white',
        'axes.facecolor': '#FCFBF7',
        'axes.edgecolor': COLORS['border'],
        'axes.labelcolor': COLORS['dark'],
        'axes.titlecolor': COLORS['dark'],
        'axes.titleweight': 'bold',
        'axes.titlesize': 14,
        'grid.color': '#E6DED1',
        'grid.linewidth': 0.8,
    },
)

RANDOM_STATE = 42
ID_COL = 'PassengerId'
TARGET = 'Transported'
SPEND_COLS = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']


def fmt_pct(value, digits=1):
    if pd.isna(value):
        return '—'
    return f'{value:.{digits}%}'


def fmt_num(value, digits=3):
    if pd.isna(value):
        return '—'
    return f'{value:,.{digits}f}'


def display_metric_cards(cards, title=None, subtitle=None):
    header = ''
    if title:
        header = f"<div style='font-size:1.35rem; font-weight:950; color:{COLORS['dark']}; margin-bottom:4px;'>{escape(title)}</div>"
    if subtitle:
        header += f"<div style='font-size:13px; color:#5F6B60; line-height:1.55; margin-bottom:14px;'>{escape(subtitle)}</div>"

    card_html = []
    for card in cards:
        color = card.get('color', COLORS['primary'])
        card_html.append(
            "<div style='background:rgba(255,255,255,0.92); border:1px solid #E4DAC9; border-radius:14px; "
            "padding:12px 14px; box-shadow:0 6px 14px rgba(35,49,39,0.045);'>"
            f"<div style='font-size:11px; text-transform:uppercase; color:#6F7C69; font-weight:900; letter-spacing:.08em; margin-bottom:6px;'>{escape(card['label'])}</div>"
            f"<div style='font-size:1.45rem; color:{color}; font-weight:950; line-height:1.1; margin-bottom:5px;'>{escape(str(card['value']))}</div>"
            f"<div style='font-size:12.5px; color:{COLORS['dark']}; line-height:1.45;'>{escape(card.get('detail', ''))}</div>"
            "</div>"
        )
    html = (
        f"<div style='background:linear-gradient(135deg, {COLORS['panel']} 0%, {COLORS['surface']} 100%); border:1px solid {COLORS['border']}; "
        "border-radius:16px; padding:18px 20px; margin:12px 0 18px 0; box-shadow:0 8px 20px rgba(35,49,39,0.04);'>"
        f"{header}<div style='display:grid; grid-template-columns:repeat(auto-fit,minmax(180px,1fr)); gap:10px;'>{''.join(card_html)}</div></div>"
    )
    display(HTML(html))


def style_dataframe(df, formats=None, bars=None, gradients=None, highlights=None):
    display_df = df.copy()
    for col in display_df.columns:
        display_df[col] = display_df[col].map(lambda value: ', '.join(map(str, value)) if isinstance(value, (list, tuple, set)) else value)

    def infer_table_formats(frame, explicit_formats=None):
        inferred = dict(explicit_formats or {})
        ratio_like_tokens = ['rate', 'share', 'ratio', 'accuracy', 'auc', 'loss', 'smd', 'correlation', 'drop', 'std', 'mean', 'threshold', 'entropy']
        for col in frame.columns:
            if col in inferred:
                continue
            numeric = pd.to_numeric(frame[col], errors='coerce')
            valid = numeric.dropna()
            if valid.empty:
                continue
            col_lower = str(col).lower()
            is_ratio_like = any(token in col_lower for token in ratio_like_tokens)
            is_integer_like = np.all(np.isclose(valid, np.round(valid), atol=1e-9))
            if is_integer_like and not is_ratio_like:
                inferred[col] = '{:,.0f}'
            elif 'rate' in col_lower or 'share' in col_lower:
                inferred[col] = '{:.2%}' if valid.between(0, 1).all() else '{:.2f}'
            elif is_ratio_like:
                inferred[col] = '{:.3f}'
            else:
                inferred[col] = '{:,.3f}'
        return inferred

    table_formats = infer_table_formats(display_df, formats)
    styler = display_df.style.format(table_formats, na_rep='—')
    styler = styler.set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#F2F3F1'), ('color', COLORS['dark']), ('font-weight', '900'), ('border', f'1px solid {COLORS["border"]}'), ('padding', '9px 10px')]},
        {'selector': 'td', 'props': [('border', '1px solid #EFE5D8'), ('padding', '8px 10px'), ('color', COLORS['dark'])]},
        {'selector': 'tbody tr:nth-child(even)', 'props': [('background-color', COLORS['table_alt'])]},
        {'selector': 'tbody tr:nth-child(odd)', 'props': [('background-color', COLORS['panel'])]},
        {'selector': 'caption', 'props': [('caption-side', 'top'), ('font-weight', '900'), ('font-size', '1.05rem'), ('color', COLORS['dark']), ('text-align', 'left'), ('padding', '0 0 8px 0')]},
    ]).set_properties(**{'font-size': '12.5px', 'line-height': '1.45'})

    gradient_cols = {col for col in (gradients or []) if col in display_df.columns}

    def relative_luminance(hex_color):
        rgb = np.array(mcolors.to_rgb(hex_color))
        linear = np.where(rgb <= 0.03928, rgb / 12.92, ((rgb + 0.055) / 1.055) ** 2.4)
        return 0.2126 * linear[0] + 0.7152 * linear[1] + 0.0722 * linear[2]

    def contrast_ratio(color_a, color_b):
        lum_a, lum_b = relative_luminance(color_a), relative_luminance(color_b)
        lighter, darker = max(lum_a, lum_b), min(lum_a, lum_b)
        return (lighter + 0.05) / (darker + 0.05)

    def contrast_text_color(hex_color):
        candidates = ['#000000', '#FFFFFF']
        return max(candidates, key=lambda color: contrast_ratio(hex_color, color))

    def gradient_contrast_styles(values):
        styles = pd.DataFrame('', index=values.index, columns=values.columns)
        cmap = plt.get_cmap('YlGnBu')
        for col in gradient_cols:
            numeric = pd.to_numeric(values[col], errors='coerce')
            valid = numeric.dropna()
            if valid.empty:
                continue
            lo, hi = valid.min(), valid.max()
            denom = hi - lo if hi != lo else 1
            normalized = ((numeric - lo) / denom).fillna(0.0).clip(0, 1)
            for idx, value in normalized.items():
                bg = mcolors.to_hex(cmap(value))
                fg = contrast_text_color(bg)
                shadow = '0 1px 1px rgba(0,0,0,0.35)' if fg == '#FFFFFF' else 'none'
                styles.loc[idx, col] = (
                    f'background-color: {bg} !important; '
                    f'color: {fg} !important; '
                    'font-weight: 950 !important; '
                    f'text-shadow: {shadow};'
                )
        return styles

    for col in bars or []:
        if col in display_df.columns:
            styler = styler.bar(subset=[col], color='#DCEBDD')
    if highlights:
        for col, color in highlights.items():
            if col in display_df.columns:
                props = {'font-weight': '900'}
                if col not in gradient_cols:
                    props['color'] = color
                styler = styler.set_properties(subset=[col], **props)
    if gradient_cols:
        styler = styler.apply(gradient_contrast_styles, axis=None)
    return styler


def display_table_card(title, df, subtitle=None, index=False, formats=None, bars=None, gradients=None, highlights=None):
    styler = style_dataframe(df, formats=formats, bars=bars, gradients=gradients, highlights=highlights)
    if not index:
        try:
            styler = styler.hide(axis='index')
        except Exception:
            pass
    table_html = styler.to_html()
    subtitle_html = f"<div style='font-size:13px; color:#5F6B60; line-height:1.55; margin-bottom:12px;'>{escape(subtitle)}</div>" if subtitle else ''
    html = (
        f"<div style='background:{COLORS['panel']}; border:1px solid {COLORS['border']}; border-radius:16px; "
        "padding:15px 16px; margin:12px 0 18px 0; box-shadow:0 8px 20px rgba(35,49,39,0.04); overflow-x:auto;'>"
        f"<div style='font-size:1.18rem; font-weight:950; color:{COLORS['dark']}; margin-bottom:5px;'>{escape(title)}</div>{subtitle_html}{table_html}</div>"
    )
    display(HTML(html))
```

```python
# =================================================================================
# DATA LOADING
# =================================================================================
def resolve_file(filename):
    roots = [
        Path('/kaggle/input/competitions/spaceship-titanic'),
        Path('/kaggle/input/spaceship-titanic'),
        Path('.'),
    ]
    for root in roots:
        path = root / filename
        if path.exists():
            return path
    raise FileNotFoundError(f'Could not locate {filename}')

train_path = resolve_file('train.csv')
test_path = resolve_file('test.csv')
submission_template_path = resolve_file('sample_submission.csv')

train = pd.read_csv(train_path)
test = pd.read_csv(test_path)
sample_submission = pd.read_csv(submission_template_path)

file_summary = pd.DataFrame({
    '📄 Table': ['train', 'test', 'sample_submission'],
    'Rows': [len(train), len(test), len(sample_submission)],
    'Columns': [train.shape[1], test.shape[1], sample_submission.shape[1]],
    'Role': ['labeled learning data', 'unlabeled prediction target', 'required submission schema'],
})

display_table_card(
    'Competition files',
    file_summary,
    subtitle='The train/test sizes are small enough for fast iteration, but large enough for stable group-aware validation.',
    index=False,
    formats={'Rows': '{:,}', 'Columns': '{:,}'},
    highlights={'Rows': COLORS['primary'], 'Columns': COLORS['signal']},
)
display_table_card('Training preview', train.head(), subtitle='Identifiers, travel context, cabin structure, spending, and target values.', index=False)
```

## 4. 🛠️ Feature Engineering

| Feature family | Source | Why it can help |
|---|---|---|
| 👥 Group size | `PassengerId` | Passengers in the same booking may share context, while raw group IDs do not generalize. |
| 🚪 Cabin deck, side, and interactions | `Cabin` | Ship location and deck-side combinations can shift the base rate. |
| 🧩 Rule-based imputations | `CryoSleep`, spending, cabin, name | Some missing values can be filled with target-free structural rules. |
| 💸 Observed spend, known no-spend, log spend, and ratios | Spending columns | Spending is powerful, but missing spend should not be over-interpreted as true zero spend. |
| 👶 Age group | `Age` | Nonlinear age patterns are easier to capture. |
| 👤 Family size | `Name` surname | Family-like structure can add context without using the target. |
| 🕳️ Missing indicators | Raw columns | Missingness itself can carry useful information. |


```python
# Build reusable feature engineering logic

def parse_identity_fields(df):
    parsed = df.copy()
    passenger_parts = parsed[ID_COL].astype('string').str.split('_', expand=True)
    parsed['Group'] = passenger_parts[0].astype('string')
    parsed['PassengerNumber'] = pd.to_numeric(passenger_parts[1], errors='coerce')

    cabin_parts = parsed['Cabin'].astype('string').str.split('/', expand=True)
    parsed['CabinDeckRaw'] = cabin_parts[0].astype('object')
    parsed['CabinNum'] = pd.to_numeric(cabin_parts[1], errors='coerce')
    parsed['CabinSideRaw'] = cabin_parts[2].astype('object')
    parsed['Surname'] = parsed['Name'].astype('string').str.split().str[-1].fillna('Unknown')
    return parsed


def most_frequent_by_key(frame, key_col, value_col):
    usable = frame[[key_col, value_col]].dropna()
    usable = usable[usable[key_col].astype(str) != 'Unknown']
    if usable.empty:
        return pd.Series(dtype='object')
    return usable.groupby(key_col)[value_col].agg(lambda values: values.mode().iat[0])


def add_spaceship_features(df, reference_df=None):
    data = parse_identity_fields(df)
    reference = parse_identity_fields(data if reference_df is None else reference_df)

    for col in ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age', 'VIP', 'Name']:
        data[f'{col}Missing'] = data[col].isna().astype(int)

    for col in SPEND_COLS:
        data[col] = pd.to_numeric(data[col], errors='coerce')
        reference[col] = pd.to_numeric(reference[col], errors='coerce')

    spend_missing_before_rules = data[SPEND_COLS].isna().sum(axis=1)
    observed_spend_before_rules = data[SPEND_COLS].fillna(0)
    any_spend_before_rules = observed_spend_before_rules.sum(axis=1).gt(0)

    group_home = most_frequent_by_key(reference, 'Group', 'HomePlanet')
    surname_home = most_frequent_by_key(reference, 'Surname', 'HomePlanet')
    deck_home_constraints = {'A': 'Europa', 'B': 'Europa', 'C': 'Europa', 'T': 'Europa', 'G': 'Earth'}

    home_from_group = data['Group'].map(group_home)
    home_from_surname = data['Surname'].map(surname_home)
    home_from_deck = data['CabinDeckRaw'].map(deck_home_constraints)
    data['HomePlanetImputedBy'] = np.select(
        [data['HomePlanet'].notna(), home_from_group.notna(), home_from_surname.notna(), home_from_deck.notna()],
        ['Observed', 'GroupMode', 'SurnameMode', 'DeckRule'],
        default='Unfilled',
    )
    data['HomePlanet'] = data['HomePlanet'].combine_first(home_from_group).combine_first(home_from_surname).combine_first(home_from_deck)

    destination_from_group = most_frequent_by_key(reference, 'Group', 'Destination')
    data['Destination'] = data['Destination'].combine_first(data['Group'].map(destination_from_group))

    data.loc[any_spend_before_rules & data['CryoSleep'].isna(), 'CryoSleep'] = False
    data.loc[data['CryoSleep'].eq(True), SPEND_COLS] = data.loc[data['CryoSleep'].eq(True), SPEND_COLS].fillna(0)
    data.loc[data['HomePlanet'].eq('Earth') & data['VIP'].isna(), 'VIP'] = False

    data['CabinDeck'] = data['CabinDeckRaw'].fillna('Unknown').astype('string')
    data['CabinSide'] = data['CabinSideRaw'].fillna('Unknown').astype('string')
    data['DeckSide'] = data['CabinDeck'].astype(str) + '_' + data['CabinSide'].astype(str)
    data['CabinRegion'] = pd.cut(
        data['CabinNum'],
        bins=[-1, 300, 600, 900, 1200, 2000],
        labels=['Front', 'FrontMid', 'Mid', 'BackMid', 'Back'],
    )
    data['CabinZone'] = data['CabinRegion'].astype('object').fillna('Unknown')
    data['CabinNumParity'] = data['CabinNum'].mod(2).fillna(-1).astype(int)
    data['CabinNumBucket'] = pd.cut(
        data['CabinNum'],
        bins=10,
        labels=False,
        duplicates='drop',
    ).fillna(-1).astype(int)

    group_size = reference['Group'].value_counts(dropna=False)
    surname_size = reference.loc[reference['Surname'] != 'Unknown', 'Surname'].value_counts(dropna=False)
    cabin_size = reference['Cabin'].value_counts(dropna=False)
    surname_group_key = reference['Group'].astype(str) + '_' + reference['Surname'].astype(str)
    surname_group_size = surname_group_key.value_counts(dropna=False)

    data['GroupSize'] = data['Group'].map(group_size).fillna(1).astype(int)
    data['SoloTraveler'] = (data['GroupSize'] == 1).astype(int)
    data['GroupSizeBand'] = pd.cut(
        data['GroupSize'],
        bins=[0, 1, 2, 4, 8],
        labels=['Solo', 'Pair', 'SmallGroup', 'LargeGroup'],
        include_lowest=True,
    )
    data['FamilySize'] = data['Surname'].map(surname_size).fillna(1).astype(int)
    data['SurnameGroupSize'] = (data['Group'].astype(str) + '_' + data['Surname'].astype(str)).map(surname_group_size).fillna(1).astype(int)
    data['FamilySizeBand'] = pd.cut(
        data['FamilySize'],
        bins=[0, 1, 2, 4, 20],
        labels=['SoloName', 'PairName', 'SmallFamilyName', 'LargeFamilyName'],
        include_lowest=True,
    )
    data['CabinSize'] = data['Cabin'].map(cabin_size).fillna(1).astype(int)

    data['SpendMissingCount'] = spend_missing_before_rules
    data['SpendMissingBeforeRules'] = spend_missing_before_rules
    data['SpendMissingAfterRules'] = data[SPEND_COLS].isna().sum(axis=1)
    data['AllSpendObserved'] = data['SpendMissingAfterRules'].eq(0).astype(int)
    spend_for_features = data[SPEND_COLS].fillna(0)
    data['ObservedTotalSpend'] = spend_for_features.sum(axis=1)
    data['AnySpendPositive'] = data['ObservedTotalSpend'].gt(0).astype(int)
    data['NoSpendKnown'] = (data['ObservedTotalSpend'].eq(0) & data['AllSpendObserved'].eq(1)).astype(int)
    data['NoSpendUnknown'] = (data['ObservedTotalSpend'].eq(0) & data['AllSpendObserved'].eq(0)).astype(int)
    data['TotalSpend'] = data['ObservedTotalSpend']
    data['LuxurySpend'] = spend_for_features[['RoomService', 'Spa', 'VRDeck']].sum(axis=1)
    data['SocialSpend'] = spend_for_features[['FoodCourt', 'ShoppingMall']].sum(axis=1)
    data['NoSpend'] = data['ObservedTotalSpend'].eq(0).astype(int)
    data['SpendCountPositive'] = spend_for_features.gt(0).sum(axis=1)

    spend_ratio = spend_for_features.div(data['TotalSpend'].where(data['TotalSpend'].gt(0), np.nan), axis=0).fillna(0).clip(0, 1)
    safe_spend_ratio = spend_ratio.replace(0, 1e-9)
    data['SpendEntropy'] = -(safe_spend_ratio * np.log(safe_spend_ratio)).sum(axis=1)
    data.loc[data['TotalSpend'].eq(0), 'SpendEntropy'] = 0.0
    data['MaxSpendShare'] = spend_ratio.max(axis=1)
    data['MaxSpendCategory'] = spend_for_features.idxmax(axis=1)
    data.loc[data['TotalSpend'].eq(0), 'MaxSpendCategory'] = 'None'
    data['SpendState'] = np.select(
        [data['AnySpendPositive'].eq(1), data['NoSpendKnown'].eq(1), data['NoSpendUnknown'].eq(1)],
        ['PositiveSpend', 'KnownZeroSpend', 'UnknownZeroSpend'],
        default='MixedMissingSpend',
    )

    data['LogTotalSpend'] = np.log1p(data['TotalSpend'])
    for col in SPEND_COLS:
        data[f'Log{col}'] = np.log1p(spend_for_features[col])
        data[f'{col}Ratio'] = spend_for_features[col] / (data['TotalSpend'] + 1)
    data['SpendPerAge'] = data['TotalSpend'] / (pd.to_numeric(data['Age'], errors='coerce') + 1)
    data['LuxurySpendShare'] = data['LuxurySpend'] / (data['TotalSpend'] + 1)
    data['SocialSpendShare'] = data['SocialSpend'] / (data['TotalSpend'] + 1)

    age = pd.to_numeric(data['Age'], errors='coerce')
    data['AgeGroup'] = pd.cut(
        age,
        bins=[-1, 12, 18, 30, 50, 100],
        labels=['Child', 'Teen', 'YoungAdult', 'Adult', 'Senior'],
    )
    data['IsChild'] = (age <= 12).astype(float)
    data['IsWorkingAge'] = ((age >= 18) & (age <= 60)).astype(float)

    hp = data['HomePlanet'].astype('object').fillna('Missing').astype(str)
    dest = data['Destination'].astype('object').fillna('Missing').astype(str)
    cryo = data['CryoSleep'].astype('object').fillna('Missing').astype(str)
    data['HomeDest'] = hp + '_' + dest
    data['DeckHome'] = data['CabinDeck'].astype(str) + '_' + hp
    data['DeckSideHome'] = data['DeckSide'].astype(str) + '_' + hp
    data['HomePlanetDestination'] = data['HomeDest']
    data['DeckHomePlanet'] = data['DeckHome']
    data['DeckSideHomePlanet'] = data['DeckSideHome']
    data['CryoSleepNoSpend'] = cryo + '_' + data['NoSpend'].astype(str)
    data['CryoNoSpendCat'] = cryo + '_' + data['SpendState'].astype(str)

    candidate_missing_cols = [col for col in data.columns if col not in [TARGET]]
    data['MissingFeatureCount'] = data[candidate_missing_cols].isna().sum(axis=1)

    return data
```

```python
# Apply feature engineering and define model columns
combined_features_source = pd.concat([train.drop(columns=[TARGET]), test], axis=0, ignore_index=True)
train_fe = add_spaceship_features(train, combined_features_source)
test_fe = add_spaceship_features(test, combined_features_source)
train_fe[TARGET] = train[TARGET].astype(int)

raw_identifier_cols = [ID_COL, 'Cabin', 'Name', 'Group', 'Surname', 'CabinDeckRaw', 'CabinSideRaw']
audit_or_duplicate_engineered_cols = ['ObservedTotalSpend', 'SpendMissingBeforeRules', 'SpendMissingAfterRules']
drop_cols = raw_identifier_cols + audit_or_duplicate_engineered_cols + [TARGET]
all_feature_cols = [col for col in train_fe.columns if col not in drop_cols and col in test_fe.columns]

PUBLIC_EXTRA_FEATURES = [
    'SpendEntropy', 'MaxSpendCategory', 'MaxSpendShare', 'CabinNumParity',
    'CabinNumBucket', 'CabinZone', 'SurnameGroupSize', 'DeckSideHome',
    'CryoNoSpendCat', 'AllSpendObserved', 'NoSpendKnown', 'NoSpendUnknown',
    'TotalSpend', 'SpendState', 'HomeDest', 'DeckHome',
]
cv_gated_public_features = [feature for feature in PUBLIC_EXTRA_FEATURES if feature in all_feature_cols]
feature_cols = [col for col in all_feature_cols if col not in cv_gated_public_features]
experimental_feature_cols = all_feature_cols

cat_cols = train_fe[feature_cols].select_dtypes(include=['object', 'string', 'category', 'bool']).columns.tolist()
num_cols = [col for col in feature_cols if col not in cat_cols]
experimental_cat_cols = train_fe[experimental_feature_cols].select_dtypes(include=['object', 'string', 'category', 'bool']).columns.tolist()
experimental_num_cols = [col for col in experimental_feature_cols if col not in experimental_cat_cols]

# OneHotEncoder expects each categorical column to contain a consistent data type.
# Observed categories become strings; missing values stay as np.nan for SimpleImputer.
for frame in [train_fe, test_fe]:
    for col in experimental_cat_cols:
        frame[col] = frame[col].astype('object')
        frame[col] = frame[col].where(frame[col].notna(), np.nan)
        frame[col] = frame[col].map(lambda value: str(value) if pd.notna(value) else np.nan)

BASIC_FEATURES = ['HomePlanet', 'CryoSleep', 'Destination', 'Age', 'VIP'] + SPEND_COLS
CABIN_FEATURES = ['CabinDeck', 'CabinNum', 'CabinSide', 'DeckSide', 'CabinRegion', 'CabinSize']
SPEND_FEATURES = [
    'LuxurySpend', 'SocialSpend', 'NoSpend', 'AnySpendPositive',
    'SpendCountPositive', 'LogTotalSpend', 'SpendPerAge', 'LuxurySpendShare',
    'SocialSpendShare',
] + [f'Log{col}' for col in SPEND_COLS] + [f'{col}Ratio' for col in SPEND_COLS]
GROUP_NAME_FEATURES = ['PassengerNumber', 'GroupSize', 'SoloTraveler', 'GroupSizeBand', 'FamilySize', 'FamilySizeBand']
INTERACTION_FEATURES = ['HomePlanetDestination', 'DeckHomePlanet', 'DeckSideHomePlanet', 'CryoSleepNoSpend', 'HomePlanetImputedBy']
MISSING_FEATURES = [
    col for col in feature_cols
    if col.endswith('Missing') or col in ['SpendMissingCount', 'MissingFeatureCount']
]

ABLATION_STEPS = [
    ('1. Raw basics', BASIC_FEATURES),
    ('2. + cabin parsing', BASIC_FEATURES + CABIN_FEATURES),
    ('3. + spending engineering', BASIC_FEATURES + CABIN_FEATURES + SPEND_FEATURES),
    ('4. + group/name counts', BASIC_FEATURES + CABIN_FEATURES + SPEND_FEATURES + GROUP_NAME_FEATURES),
    ('5. + interactions', BASIC_FEATURES + CABIN_FEATURES + SPEND_FEATURES + GROUP_NAME_FEATURES + INTERACTION_FEATURES),
    ('6. + missing indicators', feature_cols),
]

feature_inventory = pd.DataFrame({
    'Feature type': [
        'Default numeric model features',
        'Default categorical model features',
        'Engineered experimental features CV-gated from default model',
        'Raw fields excluded from model',
        'Audit or duplicate engineered fields excluded',
    ],
    'Count': [len(num_cols), len(cat_cols), len(cv_gated_public_features), len(raw_identifier_cols), len(audit_or_duplicate_engineered_cols)],
    'Fields': [
        ', '.join(num_cols),
        ', '.join(cat_cols),
        ', '.join(cv_gated_public_features),
        ', '.join(raw_identifier_cols),
        ', '.join(audit_or_duplicate_engineered_cols),
    ],
})

public_feature_status = pd.DataFrame({
    'Feature': PUBLIC_EXTRA_FEATURES,
    'Engineered': [feature in all_feature_cols for feature in PUBLIC_EXTRA_FEATURES],
    'Default model': [feature in feature_cols for feature in PUBLIC_EXTRA_FEATURES],
    'Experimental candidates': [feature in experimental_feature_cols for feature in PUBLIC_EXTRA_FEATURES],
})

display_table_card(
    'Model feature inventory',
    feature_inventory,
    subtitle='Target-free extra features are available for experimental candidates; the default model keeps the CV-stable core set.',
    index=False,
    formats={'Count': '{:,}'},
    highlights={'Count': COLORS['primary']},
)
display_table_card(
    'Clean extra feature additions',
    public_feature_status,
    subtitle='Target-free extra features, public overrides excluded.',
    index=False,
)
display_table_card('Engineered feature preview', train_fe[feature_cols + [TARGET]].head(), index=False)
```

**Why raw identifiers are excluded:** `PassengerId`, `Group`, `Name`, and the full `Cabin` string are too specific. Train and test passenger groups do not overlap, so the raw group ID is not useful for generalization. We keep the useful structure through member number, group size, surname counts, deck, cabin number, side, and target-free imputation signals.

```python
# Compute target rates for important categorical features
base_rate = train_fe[TARGET].mean()
categorical_focus = ['CryoSleep', 'HomePlanet', 'Destination', 'VIP', 'CabinDeck', 'CabinSide', 'DeckSide', 'CabinRegion', 'AgeGroup', 'GroupSizeBand', 'FamilySizeBand', 'CryoSleepNoSpend']
rate_tables = []

for col in categorical_focus:
    temp = train_fe[[col, TARGET]].copy()
    temp[col] = temp[col].astype('object').fillna('Missing')
    summary = (
        temp.groupby(col, observed=True)[TARGET]
        .agg(Rows='count', TransportedRate='mean')
        .reset_index()
        .rename(columns={col: 'Category'})
    )
    summary.insert(0, 'Feature', col)
    summary['LiftVsBase'] = summary['TransportedRate'] / base_rate
    rate_tables.append(summary)

categorical_rates = pd.concat(rate_tables, ignore_index=True)
categorical_rates_display = (
    categorical_rates
    .sort_values(['Feature', 'TransportedRate'], ascending=[True, False])
    .rename(columns={'TransportedRate': 'Transported rate', 'LiftVsBase': 'Lift vs base'})
)

display_table_card(
    'Categorical target-rate priors',
    categorical_rates_display,
    subtitle='Bars far from the global base rate are useful priors, especially when support is not tiny.',
    index=False,
    formats={'Rows': '{:,}', 'Transported rate': '{:.2%}', 'Lift vs base': '{:.2f}'},
    gradients=['Transported rate'],
    highlights={'Transported rate': COLORS['risk'], 'Lift vs base': COLORS['signal']},
)
```

## 6. ⚖️ Train/Test Similarity

Validation works best when train and test come from similar distributions. We cannot see the test labels, but we can compare feature distributions. Large train/test differences are a warning that local validation may not match the Kaggle score perfectly.

```python
# Estimate numeric train/test drift with standardized mean difference

def standardized_mean_difference(left, right):
    left = pd.to_numeric(left, errors='coerce')
    right = pd.to_numeric(right, errors='coerce')
    pooled = np.sqrt((left.var(skipna=True) + right.var(skipna=True)) / 2)
    if pd.isna(pooled) or pooled == 0:
        return 0.0
    return (left.mean(skipna=True) - right.mean(skipna=True)) / pooled

numeric_drift_rows = []
for col in num_cols:
    numeric_drift_rows.append({
        'Feature': col,
        'TrainMean': pd.to_numeric(train_fe[col], errors='coerce').mean(),
        'TestMean': pd.to_numeric(test_fe[col], errors='coerce').mean(),
        'SMD': standardized_mean_difference(train_fe[col], test_fe[col]),
    })

numeric_drift = pd.DataFrame(numeric_drift_rows)
numeric_drift['AbsSMD'] = numeric_drift['SMD'].abs()

display_table_card(
    'Numeric train/test drift',
    numeric_drift.sort_values('AbsSMD', ascending=False).head(15),
    subtitle='Standardized mean difference near zero means train and test are numerically similar for that feature.',
    index=False,
    formats={'TrainMean': '{:.3f}', 'TestMean': '{:.3f}', 'SMD': '{:.3f}', 'AbsSMD': '{:.3f}'},
    bars=['AbsSMD'],
    highlights={'AbsSMD': COLORS['accent']},
)

fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=numeric_drift.sort_values('AbsSMD', ascending=False).head(12), x='AbsSMD', y='Feature', color=COLORS['signal'], ax=ax)
ax.set_title('Largest Numeric Train/Test Distribution Differences')
ax.set_xlabel('|standardized mean difference|')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Estimate categorical train/test drift with maximum category share difference
cat_drift_rows = []
for col in cat_cols:
    train_share = train_fe[col].astype('object').fillna('Missing').value_counts(normalize=True)
    test_share = test_fe[col].astype('object').fillna('Missing').value_counts(normalize=True)
    all_levels = train_share.index.union(test_share.index)
    diff = (train_share.reindex(all_levels, fill_value=0) - test_share.reindex(all_levels, fill_value=0)).abs()
    cat_drift_rows.append({
        'Feature': col,
        'MaxShareDifference': diff.max(),
        'MostDifferentCategory': diff.idxmax(),
    })

cat_drift = pd.DataFrame(cat_drift_rows).sort_values('MaxShareDifference', ascending=False)
display_table_card(
    'Categorical train/test drift',
    cat_drift,
    subtitle='This checks whether a category has a noticeably different share in test than in train.',
    index=False,
    formats={'MaxShareDifference': '{:.2%}'},
    bars=['MaxShareDifference'],
    highlights={'MaxShareDifference': COLORS['accent']},
)
```

## 7. 🤖 Modeling Plan

Group-aware validation keeps passengers from the same `PassengerId` group inside the same fold. `StratifiedGroupKFold` preserves class balance while reducing group leakage.

| Model family | Role |
|---|---|
| Logistic Regression | Linear baseline. |
| Random Forest / Extra Trees | Nonlinear tree baselines. |
| HistGradientBoosting | Strong scikit-learn tabular baseline. |
| Soft Voting Ensemble | Averaged scikit-learn probabilities. |
| CatBoost | Categorical interactions and raw surname, GPU-ready on Kaggle P100. |
| LightGBM + fold-safe target encoding | Compact categorical encoding without full-train target leakage. |
| OOF blend | Transparent probability-weight search using out-of-fold predictions. |

<div style="background:#FFFCF8; border:1px solid #E2D8C7; border-left:6px solid #3D6F8A; border-radius:14px; padding:13px 15px; margin:12px 0 0 0; color:#233127; line-height:1.58; box-shadow:0 8px 18px rgba(35,49,39,0.04);"><b>CV note:</b> count features use target-free train+test covariates such as surname and cabin counts. The score is therefore closer to a leaderboard-style transductive estimate than a fully strict future-data estimate.</div>


```python
# Create one group-aware validation split
X = train_fe[feature_cols].copy()
y = train_fe[TARGET].astype(int)
X_test = test_fe[feature_cols].copy()
groups = train_fe['Group'].astype(str)

group_cv = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
train_idx, valid_idx = next(group_cv.split(X, y, groups=groups))

X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]
groups_train, groups_valid = groups.iloc[train_idx], groups.iloc[valid_idx]

split_summary = pd.DataFrame({
    'Split': ['Training fold', 'Validation fold'],
    'Rows': [len(X_train), len(X_valid)],
    'Unique groups': [groups_train.nunique(), groups_valid.nunique()],
    'Positive rate': [y_train.mean(), y_valid.mean()],
})

display_table_card(
    'One group-aware holdout split',
    split_summary,
    subtitle='Every passenger group is kept entirely inside either training or validation for this diagnostic split.',
    index=False,
    formats={'Rows': '{:,}', 'Unique groups': '{:,}', 'Positive rate': '{:.2%}'},
    highlights={'Positive rate': COLORS['risk'], 'Unique groups': COLORS['signal']},
)
```

```python
# Define preprocessing and model pipelines

def make_one_hot_encoder():
    params = {'handle_unknown': 'ignore'}
    if 'sparse_output' in inspect.signature(OneHotEncoder).parameters:
        params['sparse_output'] = False
    else:
        params['sparse'] = False
    return OneHotEncoder(**params)


def split_feature_types(columns):
    local_cat_cols = [col for col in columns if col in cat_cols]
    local_num_cols = [col for col in columns if col in num_cols]
    return local_num_cols, local_cat_cols


def build_preprocessor(columns=None):
    columns = feature_cols if columns is None else list(columns)
    local_num_cols, local_cat_cols = split_feature_types(columns)
    numeric_pipe = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median', add_indicator=True)),
        ('scaler', StandardScaler()),
    ])
    categorical_pipe = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='Missing')),
        ('onehot', make_one_hot_encoder()),
    ])
    return ColumnTransformer(
        transformers=[
            ('num', numeric_pipe, local_num_cols),
            ('cat', categorical_pipe, local_cat_cols),
        ],
        remainder='drop',
        verbose_feature_names_out=True,
    )


def make_random_forest():
    return RandomForestClassifier(
        n_estimators=200,
        min_samples_leaf=3,
        max_features='sqrt',
        n_jobs=-1,
        random_state=RANDOM_STATE,
    )


def make_extra_trees():
    return ExtraTreesClassifier(
        n_estimators=250,
        min_samples_leaf=3,
        max_features='sqrt',
        n_jobs=-1,
        random_state=RANDOM_STATE,
    )


def make_hist_gradient_boosting():
    return HistGradientBoostingClassifier(
        learning_rate=0.045,
        max_iter=260,
        max_leaf_nodes=24,
        l2_regularization=0.04,
        random_state=RANDOM_STATE,
    )

models = {
    'Logistic Regression': LogisticRegression(max_iter=2500, C=1.0, random_state=RANDOM_STATE),
    'Random Forest': make_random_forest(),
    'Extra Trees': make_extra_trees(),
    'HistGradientBoosting': make_hist_gradient_boosting(),
    'Soft Voting Ensemble': VotingClassifier(
        estimators=[
            ('rf', make_random_forest()),
            ('et', make_extra_trees()),
            ('hgb', make_hist_gradient_boosting()),
        ],
        voting='soft',
        weights=[2, 1, 2],
        n_jobs=1,
    ),
}


def build_pipeline(model, columns=None):
    return Pipeline(steps=[
        ('preprocess', build_preprocessor(columns)),
        ('model', clone(model)),
    ])
```

```python
# Fit baselines and evaluate holdout validation performance
fitted_models = {}
valid_predictions = {}
holdout_rows = []

for name, model in models.items():
    pipe = build_pipeline(model)
    pipe.fit(X_train, y_train)
    proba = pipe.predict_proba(X_valid)[:, 1]
    pred = (proba >= 0.5).astype(int)

    fitted_models[name] = pipe
    valid_predictions[name] = {'proba': proba, 'pred': pred}
    holdout_rows.append({
        'Model': name,
        'Accuracy': accuracy_score(y_valid, pred),
        'ROC AUC': roc_auc_score(y_valid, proba),
        'Log loss': log_loss(y_valid, proba, labels=[0, 1]),
    })

holdout_leaderboard = (
    pd.DataFrame(holdout_rows)
    .sort_values(['Accuracy', 'ROC AUC'], ascending=[False, False])
    .reset_index(drop=True)
)

display_table_card(
    'Holdout leaderboard',
    holdout_leaderboard,
    subtitle='A quick diagnostic split. The group-CV leaderboard below is the more reliable selection evidence.',
    index=False,
    formats={'Accuracy': '{:.4f}', 'ROC AUC': '{:.4f}', 'Log loss': '{:.4f}'},
    gradients=['Accuracy'],
    highlights={'Accuracy': COLORS['risk'], 'ROC AUC': COLORS['signal']},
)
```

Accuracy measures class-label correctness at threshold `0.5`. ROC AUC measures probability ranking. Log loss penalizes confident wrong probabilities.

### 7.2 CatBoost GPU Candidate

CatBoost handles categorical features such as `DeckSide`, interaction columns, and high-cardinality surname values. On Kaggle P100, the CatBoost candidate uses GPU training with early stopping.


```python
# CatBoost OOF/test candidate for categorical-heavy tabular modeling
try:
    from catboost import CatBoostClassifier, CatBoostError
    CATBOOST_AVAILABLE = True
except ImportError:
    CATBOOST_AVAILABLE = False
    CatBoostError = RuntimeError

CATBOOST_FINAL_AVAILABLE = False
catboost_oof_proba = None
catboost_test_proba = None
catboost_results = pd.DataFrame()
CATBOOST_TASK_TYPE = 'GPU'
CATBOOST_ITERATIONS = 3200
CATBOOST_LEARNING_RATE = 0.025
CATBOOST_OD_WAIT = 180

if not CATBOOST_AVAILABLE:
    display_metric_cards(
        [{'label': '🐈 CatBoost status', 'value': 'Not installed', 'detail': 'The scikit-learn candidates remain fully usable.', 'color': COLORS['neutral']}],
        title='CatBoost Final Candidate',
    )
else:
    catboost_feature_cols = [col for col in experimental_feature_cols + ['Surname'] if col in train_fe.columns and col in test_fe.columns]
    X_cb = train_fe[catboost_feature_cols].copy()
    X_test_cb = test_fe[catboost_feature_cols].copy()
    catboost_cat_cols = [col for col in catboost_feature_cols if col in experimental_cat_cols or col == 'Surname']
    for frame in [X_cb, X_test_cb]:
        for col in catboost_cat_cols:
            frame[col] = frame[col].astype('object').where(frame[col].notna(), 'Missing').astype(str)

    catboost_cv = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    catboost_rows = []
    cat_feature_indices = [X_cb.columns.get_loc(col) for col in catboost_cat_cols]
    catboost_oof_proba = np.zeros(len(X_cb))
    catboost_test_proba = np.zeros(len(X_test_cb))
    effective_catboost_task_type = CATBOOST_TASK_TYPE

    for fold, (tr_idx, va_idx) in enumerate(catboost_cv.split(X_cb, y, groups=groups), start=1):
        model_params = {
            'loss_function': 'Logloss',
            'eval_metric': 'Accuracy',
            'iterations': CATBOOST_ITERATIONS,
            'learning_rate': CATBOOST_LEARNING_RATE,
            'depth': 6,
            'l2_leaf_reg': 8,
            'random_seed': RANDOM_STATE + fold,
            'od_type': 'Iter',
            'od_wait': CATBOOST_OD_WAIT,
            'allow_writing_files': False,
            'verbose': False,
            'task_type': effective_catboost_task_type,
        }
        if effective_catboost_task_type == 'GPU':
            model_params['devices'] = '0'

        try:
            model = CatBoostClassifier(**model_params)
            model.fit(
                X_cb.iloc[tr_idx],
                y.iloc[tr_idx],
                cat_features=cat_feature_indices,
                eval_set=(X_cb.iloc[va_idx], y.iloc[va_idx]),
            )
        except CatBoostError as error:
            if effective_catboost_task_type != 'GPU':
                raise
            effective_catboost_task_type = 'CPU'
            model_params.pop('devices', None)
            model_params['task_type'] = 'CPU'
            model = CatBoostClassifier(**model_params)
            model.fit(
                X_cb.iloc[tr_idx],
                y.iloc[tr_idx],
                cat_features=cat_feature_indices,
                eval_set=(X_cb.iloc[va_idx], y.iloc[va_idx]),
            )

        proba = model.predict_proba(X_cb.iloc[va_idx])[:, 1]
        catboost_oof_proba[va_idx] = proba
        catboost_test_proba += model.predict_proba(X_test_cb)[:, 1] / catboost_cv.get_n_splits()
        pred = (proba >= 0.5).astype(int)
        best_iteration = model.get_best_iteration()
        if best_iteration is None:
            best_iteration = model.tree_count_
        catboost_rows.append({
            'Fold': fold,
            'Task type': effective_catboost_task_type,
            'Accuracy': accuracy_score(y.iloc[va_idx], pred),
            'ROC AUC': roc_auc_score(y.iloc[va_idx], proba),
            'Best iteration': best_iteration,
        })

    catboost_results = pd.DataFrame(catboost_rows)
    catboost_mean = catboost_results['Accuracy'].mean()
    catboost_oof_acc_05 = accuracy_score(y, catboost_oof_proba >= 0.5)
    catboost_oof_auc = roc_auc_score(y, catboost_oof_proba)
    CATBOOST_FINAL_AVAILABLE = True
    display_metric_cards(
        [
            {'label': '🐈 CatBoost OOF accuracy', 'value': f'{catboost_oof_acc_05:.4f}', 'detail': 'Threshold: 0.50', 'color': COLORS['risk']},
            {'label': '⚙️ CatBoost task', 'value': effective_catboost_task_type, 'detail': f'{CATBOOST_ITERATIONS:,} max iterations, early stopping {CATBOOST_OD_WAIT}', 'color': COLORS['accent']},
            {'label': '📈 CatBoost OOF ROC AUC', 'value': f'{catboost_oof_auc:.4f}', 'detail': 'All 5 group folds', 'color': COLORS['signal']},
            {'label': '🧩 CatBoost features', 'value': f'{len(catboost_feature_cols):,}', 'detail': f'Categorical: {len(catboost_cat_cols):,}', 'color': COLORS['primary']},
        ],
        title='CatBoost Final Candidate',
    )
    display_table_card(
        'CatBoost fold results',
        catboost_results,
        index=False,
        formats={'Accuracy': '{:.4f}', 'ROC AUC': '{:.4f}', 'Best iteration': '{:,.0f}'},
        gradients=['Accuracy'],
        highlights={'Accuracy': COLORS['risk'], 'ROC AUC': COLORS['signal']},
    )
    print(f"CatBoost task type used: {effective_catboost_task_type}")
    print(f"CatBoost max iterations: {CATBOOST_ITERATIONS}")
    print(f"CatBoost mean accuracy: {catboost_mean:.4f}")
    print(f"CatBoost OOF accuracy at 0.5: {catboost_oof_acc_05:.4f}")
```

### 7.3 LightGBM with Fold-Safe Target Encoding

Target encoding can help tree models use categorical interactions compactly, but it must be learned inside each fold. For every validation fold below, category statistics are computed from the training folds only, then applied to the held-out fold and test set.


```python
# LightGBM OOF/test candidate with fold-safe target encoding
try:
    import lightgbm as lgb
    LIGHTGBM_AVAILABLE = True
except ImportError:
    LIGHTGBM_AVAILABLE = False

LIGHTGBM_FINAL_AVAILABLE = False
lightgbm_oof_proba = None
lightgbm_test_proba = None
lightgbm_results = pd.DataFrame()

lightgbm_feature_cols = experimental_feature_cols
X_lgbm = train_fe[lightgbm_feature_cols].copy()
X_test_lgbm = test_fe[lightgbm_feature_cols].copy()
lightgbm_cat_cols = [col for col in lightgbm_feature_cols if col in experimental_cat_cols]
lightgbm_num_cols = [col for col in lightgbm_feature_cols if col not in lightgbm_cat_cols]

TE_COLS = [
    'DeckSide', 'HomeDest', 'DeckHome', 'DeckSideHome', 'CabinZone',
    'AgeGroup', 'MaxSpendCategory', 'CryoSleep', 'CryoNoSpendCat',
]
TE_COLS = [col for col in TE_COLS if col in lightgbm_feature_cols]


def clean_category(series):
    return series.astype('object').where(series.notna(), 'Missing').astype(str)


def make_target_encoded_frame(train_part, apply_part, y_train_part, te_cols, smoothing=35):
    frame = pd.DataFrame(index=apply_part.index)
    for col in lightgbm_num_cols:
        if col in apply_part.columns:
            frame[col] = pd.to_numeric(apply_part[col], errors='coerce')

    global_mean = float(y_train_part.mean())
    for col in te_cols:
        train_key = clean_category(train_part[col])
        apply_key = clean_category(apply_part[col])
        stats = (
            pd.DataFrame({'category': train_key, 'target': y_train_part.to_numpy()})
            .groupby('category')['target']
            .agg(['mean', 'count'])
        )
        encoded = (stats['count'] * stats['mean'] + smoothing * global_mean) / (stats['count'] + smoothing)
        frame[f'{col}_TE'] = apply_key.map(encoded).fillna(global_mean).astype(float)

    frequency_cols = [col for col in lightgbm_cat_cols if col not in te_cols]
    for col in frequency_cols:
        train_key = clean_category(train_part[col])
        apply_key = clean_category(apply_part[col])
        frequencies = train_key.value_counts(normalize=True)
        frame[f'{col}_Freq'] = apply_key.map(frequencies).fillna(0).astype(float)

    return frame


def make_lightgbm_model(seed):
    return lgb.LGBMClassifier(
        objective='binary',
        n_estimators=1200,
        learning_rate=0.025,
        num_leaves=31,
        min_child_samples=25,
        subsample=0.85,
        colsample_bytree=0.85,
        reg_alpha=0.1,
        reg_lambda=1.5,
        random_state=seed,
        n_jobs=-1,
        verbose=-1,
    )


if not LIGHTGBM_AVAILABLE:
    display_metric_cards(
        [{'label': '💡 LightGBM status', 'value': 'Not installed', 'detail': 'The other candidates remain fully usable.', 'color': COLORS['neutral']}],
        title='LightGBM Target-Encoding Candidate',
    )
else:
    lightgbm_cv = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    lightgbm_rows = []
    lightgbm_oof_proba = np.zeros(len(X))
    lightgbm_test_proba = np.zeros(len(X_test))

    for fold, (tr_idx, va_idx) in enumerate(lightgbm_cv.split(X_lgbm, y, groups=groups), start=1):
        X_tr, X_va = X_lgbm.iloc[tr_idx], X_lgbm.iloc[va_idx]
        y_tr, y_va = y.iloc[tr_idx], y.iloc[va_idx]
        X_tr_lgb = make_target_encoded_frame(X_tr, X_tr, y_tr, TE_COLS)
        X_va_lgb = make_target_encoded_frame(X_tr, X_va, y_tr, TE_COLS)
        X_test_lgb = make_target_encoded_frame(X_tr, X_test_lgbm, y_tr, TE_COLS)

        model = make_lightgbm_model(RANDOM_STATE + fold)
        model.fit(
            X_tr_lgb,
            y_tr,
            eval_set=[(X_va_lgb, y_va)],
            eval_metric='binary_logloss',
            callbacks=[lgb.early_stopping(100, verbose=False)],
        )

        proba = model.predict_proba(X_va_lgb)[:, 1]
        lightgbm_oof_proba[va_idx] = proba
        lightgbm_test_proba += model.predict_proba(X_test_lgb)[:, 1] / lightgbm_cv.get_n_splits()
        lightgbm_rows.append({
            'Fold': fold,
            'Accuracy': accuracy_score(y_va, proba >= 0.5),
            'ROC AUC': roc_auc_score(y_va, proba),
            'Best iteration': getattr(model, 'best_iteration_', model.n_estimators),
        })

    lightgbm_results = pd.DataFrame(lightgbm_rows)
    lightgbm_oof_acc_05 = accuracy_score(y, lightgbm_oof_proba >= 0.5)
    lightgbm_oof_auc = roc_auc_score(y, lightgbm_oof_proba)
    LIGHTGBM_FINAL_AVAILABLE = True

    display_metric_cards(
        [
            {'label': '💡 LightGBM OOF accuracy', 'value': f'{lightgbm_oof_acc_05:.4f}', 'detail': 'Threshold: 0.50', 'color': COLORS['risk']},
            {'label': '📈 LightGBM OOF ROC AUC', 'value': f'{lightgbm_oof_auc:.4f}', 'detail': 'Fold-safe target encoding', 'color': COLORS['signal']},
            {'label': '🧬 TE features', 'value': f'{len(TE_COLS):,}', 'detail': ', '.join(TE_COLS), 'color': COLORS['primary']},
        ],
        title='LightGBM Target-Encoding Candidate',
    )
    display_table_card(
        'LightGBM fold results',
        lightgbm_results,
        index=False,
        formats={'Accuracy': '{:.4f}', 'ROC AUC': '{:.4f}', 'Best iteration': '{:,.0f}'},
        gradients=['Accuracy'],
        highlights={'Accuracy': COLORS['risk'], 'ROC AUC': COLORS['signal']},
    )
    print(f"LightGBM mean accuracy: {lightgbm_results['Accuracy'].mean():.4f}")
    print(f"LightGBM OOF accuracy at 0.5: {lightgbm_oof_acc_05:.4f}")
```

## 8. 🎚️ Threshold Tuning

Most classifiers output probabilities. A threshold converts probabilities into `True` or `False` labels. The default threshold is `0.5`, but accuracy can sometimes improve with a nearby threshold.

This tuning uses out-of-fold predictions from `StratifiedGroupKFold`, so every training row is scored by a model that did not train on its passenger group. That gives a steadier threshold estimate than tuning on a single holdout fold.

```python
# Tune probability thresholds with group-aware OOF predictions for final candidates
from itertools import product

selected_holdout_model = fitted_models[best_model_name]


def get_oof_and_test_predictions(model, X, y, X_test, groups, cv, columns=None):
    oof = np.zeros(len(X))
    test_pred = np.zeros(len(X_test))
    n_splits = cv.get_n_splits()

    for tr_idx, va_idx in cv.split(X, y, groups=groups):
        fold_model = build_pipeline(model, columns=columns)
        fold_model.fit(X.iloc[tr_idx], y.iloc[tr_idx])
        oof[va_idx] = fold_model.predict_proba(X.iloc[va_idx])[:, 1]
        test_pred += fold_model.predict_proba(X_test)[:, 1] / n_splits

    return oof, test_pred


def best_threshold_for_proba(y_true, proba, grid):
    accuracies = np.array([accuracy_score(y_true, proba >= threshold) for threshold in grid])
    best_idx = accuracies.argmax()
    return float(grid[best_idx]), float(accuracies[best_idx]), accuracies


sklearn_oof_proba, sklearn_test_proba = get_oof_and_test_predictions(
    models[best_model_name],
    X,
    y,
    X_test,
    groups,
    group_cv,
)

prediction_candidates = [
    {
        'Candidate': f'Scikit-learn {best_model_name}',
        'OOF proba': sklearn_oof_proba,
        'Test proba': sklearn_test_proba,
    }
]
blend_sources = [
    {
        'Short name': 'sklearn',
        'OOF proba': sklearn_oof_proba,
        'Test proba': sklearn_test_proba,
    }
]

if CATBOOST_FINAL_AVAILABLE:
    prediction_candidates.append({
        'Candidate': 'CatBoost',
        'OOF proba': catboost_oof_proba,
        'Test proba': catboost_test_proba,
    })
    blend_sources.append({
        'Short name': 'cat',
        'OOF proba': catboost_oof_proba,
        'Test proba': catboost_test_proba,
    })

if LIGHTGBM_FINAL_AVAILABLE:
    prediction_candidates.append({
        'Candidate': 'LightGBM with fold-safe TE',
        'OOF proba': lightgbm_oof_proba,
        'Test proba': lightgbm_test_proba,
    })
    blend_sources.append({
        'Short name': 'lgbm_te',
        'OOF proba': lightgbm_oof_proba,
        'Test proba': lightgbm_test_proba,
    })

threshold_grid = np.round(np.arange(0.35, 0.651, 0.001), 3)

if len(blend_sources) >= 2:
    blend_best = None
    for raw_weights in product(range(11), repeat=len(blend_sources)):
        if sum(raw_weights) != 10:
            continue
        weights = np.array(raw_weights, dtype=float) / 10
        oof_blend = sum(weight * source['OOF proba'] for weight, source in zip(weights, blend_sources))
        threshold, score, _ = best_threshold_for_proba(y, oof_blend, threshold_grid)
        if blend_best is None or score > blend_best['Score']:
            blend_best = {
                'Weights': weights,
                'Threshold': threshold,
                'Score': score,
                'OOF proba': oof_blend,
                'Test proba': sum(weight * source['Test proba'] for weight, source in zip(weights, blend_sources)),
            }

    weight_label = ' + '.join(
        f"{weight:.0%} {source['Short name']}"
        for weight, source in zip(blend_best['Weights'], blend_sources)
        if weight > 0
    )
    prediction_candidates.append({
        'Candidate': f'OOF weight blend ({weight_label})',
        'OOF proba': blend_best['OOF proba'],
        'Test proba': blend_best['Test proba'],
    })

candidate_rows = []
threshold_curve_rows = []

for candidate in prediction_candidates:
    name = candidate['Candidate']
    proba = candidate['OOF proba']
    best_candidate_threshold, best_candidate_accuracy, accuracies = best_threshold_for_proba(y, proba, threshold_grid)

    candidate_rows.append({
        'Candidate': name,
        'Accuracy at 0.50': accuracy_score(y, proba >= 0.5),
        'Best threshold': best_candidate_threshold,
        'Best OOF accuracy': best_candidate_accuracy,
        'OOF ROC AUC': roc_auc_score(y, proba),
    })
    threshold_curve_rows.extend(
        {'Candidate': name, 'Threshold': threshold, 'Accuracy': accuracy}
        for threshold, accuracy in zip(threshold_grid, accuracies)
    )

candidate_thresholds = pd.DataFrame(candidate_rows).sort_values(
    ['Best OOF accuracy', 'OOF ROC AUC'],
    ascending=[False, False],
).reset_index(drop=True)
threshold_curves = pd.DataFrame(threshold_curve_rows)

final_candidate_name = candidate_thresholds.loc[0, 'Candidate']
best_threshold = float(candidate_thresholds.loc[0, 'Best threshold'])
best_threshold_row = candidate_thresholds.iloc[0]
final_prediction_candidate = next(item for item in prediction_candidates if item['Candidate'] == final_candidate_name)
oof_proba = final_prediction_candidate['OOF proba']
final_test_proba = final_prediction_candidate['Test proba']
test_proba = final_test_proba
threshold_scores = threshold_curves[threshold_curves['Candidate'].eq(final_candidate_name)].copy()

display_metric_cards(
    [
        {'label': '🏁 Final candidate', 'value': final_candidate_name, 'detail': 'Chosen by best group-OOF accuracy', 'color': COLORS['primary']},
        {'label': '🎚️ Best OOF threshold', 'value': f'{best_threshold:.3f}', 'detail': 'Grid: 0.35 to 0.65', 'color': COLORS['signal']},
        {'label': '✅ Best OOF accuracy', 'value': f"{best_threshold_row['Best OOF accuracy']:.4f}", 'detail': f"Accuracy at 0.50: {best_threshold_row['Accuracy at 0.50']:.4f}", 'color': COLORS['risk']},
    ],
    title='Final Candidate and Threshold Selection',
    subtitle='OOF predictions compare scikit-learn, CatBoost, LightGBM target encoding, and a transparent weight-search blend.',
)
display_table_card(
    'OOF candidate comparison',
    candidate_thresholds,
    index=False,
    formats={'Accuracy at 0.50': '{:.4f}', 'Best threshold': '{:.3f}', 'Best OOF accuracy': '{:.4f}', 'OOF ROC AUC': '{:.4f}'},
    gradients=['Best OOF accuracy'],
    highlights={'Best OOF accuracy': COLORS['risk'], 'OOF ROC AUC': COLORS['signal']},
)

fig, ax = plt.subplots(figsize=(9, 4.5))
sns.lineplot(data=threshold_curves, x='Threshold', y='Accuracy', hue='Candidate', ax=ax)
ax.axvline(best_threshold, color=COLORS['risk'], linestyle='--', linewidth=1)
ax.set_title('OOF Threshold Tuning by Candidate')
ax.set_xlabel('Probability threshold')
ax.set_ylabel('OOF validation accuracy')
plt.tight_layout()
plt.show()

print(f'Final prediction candidate: {final_candidate_name}')
print(f'Best OOF threshold: {best_threshold:.3f}')
print(f'Best OOF accuracy: {best_threshold_row["Best OOF accuracy"]:.4f}')
```

## 11. ✅ Statistically Grounded Next Experiments

A useful next experiment must answer one of two questions: **does the model improve beyond validation noise, or does it explain a stable error pattern?** Small OOF differences are easy to over-read. On 8,693 training rows, an accuracy change of **0.001** is only about **9 passengers**, so candidate changes need group-aware uncertainty checks.

| Priority | Analysis | Statistical logic | Decision rule |
|---:|---|---|---|
| 1 | Group bootstrap confidence intervals for OOF candidates | Resample passenger groups, not individual rows, then recompute accuracy, ROC AUC, and log loss. This estimates how much the score changes under group-level sampling noise. | Keep a blend only if its bootstrap distribution beats the simpler model by more than noise, or if it improves log loss/AUC while matching accuracy. |
| 2 | Pairwise OOF disagreement test | Compare two candidates on the same OOF rows. The useful rows are where one model is correct and the other is wrong. McNemar-style counts show whether the win is asymmetric or just random row swapping. | If the win/loss table is nearly symmetric, prefer the simpler or better-calibrated candidate. |
| 3 | Threshold stability across folds and seeds | Accuracy is discontinuous around the threshold. Estimate the best threshold per fold/seed and inspect its spread. | Use a single global threshold only if fold thresholds cluster tightly; otherwise use median threshold or optimize for stable positive rate. |
| 4 | Strict versus transductive feature audit | Current count features can use target-free train+test covariates. A strict version should rebuild group, surname, and cabin counts using training folds only. | Treat the difference between strict CV and transductive CV as an optimism estimate. Large gaps mean leaderboard risk. |
| 5 | Fold-safe target encoding smoothing search | For interactions such as deck-side-home and max-spend category, tune smoothing strength λ. Rare categories need stronger shrinkage toward the global mean. | Select λ by OOF log loss first, accuracy second. Overfit encodings often raise accuracy in one fold but worsen log loss. |
| 6 | Calibration and reliability curves | Accuracy only checks labels; submission thresholds depend on probability quality. Use Brier score, log loss, and reliability bins. | Calibrate only when OOF calibration improves log loss without moving accuracy down. |
| 7 | Residual slice analysis | Split OOF errors by CryoSleep, no-spend status, deck, side, home planet, age band, and group size. Error concentration points to missing interactions or bad imputation assumptions. | Add features only for slices with enough rows and a repeatable error gap across folds. |
| 8 | Soft post-processing from OOF evidence | Rules such as CryoSleep + zero spend are strong but not deterministic. Estimate post-processing on OOF probabilities before touching test predictions. | Apply a bump only near the threshold and only if OOF accuracy and positive-rate stability both improve. |
| 9 | Seed ensemble variance | Refit CatBoost/HGB with several seeds and measure prediction variance. High-variance rows are usually threshold-sensitive. | Average seeds when variance reduction improves OOF log loss or stabilizes threshold choice. |

The most informative first follow-up is therefore:

1. build group-bootstrap confidence intervals for the current HGB, CatBoost, LightGBM, and OOF-blend probabilities;
2. run pairwise OOF disagreement tables against the HGB baseline;
3. inspect residual slices for CryoSleep/no-spend, deck-side-home, and group-size bands;
4. only then add XGBoost or extra post-processing.
