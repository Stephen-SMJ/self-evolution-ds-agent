# Preprocess csv file & create folder and copy files

### Load csv file

```python
import pandas as pd
csv_path = '/kaggle/input/rsna-process-dataset-jitshil/process_dataset_final.csv'
df = pd.read_csv(csv_path)
df.head()
```

### Create patient_L/R folders and copy dicom files between two folders respectively L and R letarelity

```python
import os
import shutil

source_main_path = '/kaggle/input/rsna-breast-cancer-detection/train_images/'
output_train_folder = '/kaggle/working/train_imgs/'
if os.path.exists(output_train_folder) == False:
    os.mkdir(output_train_folder)


for dfvalue in  df.values:
    folder_name = str(dfvalue[2])+'_'+ str(dfvalue[4])
    #print(folder_name)
    output_folder = '/kaggle/working/train_imgs/'+folder_name
    if os.path.exists(output_folder) == False:
        os.mkdir(output_folder)
    src = source_main_path+str(dfvalue[2])+"/"+str(dfvalue[3])+'.dcm'
    dst = output_folder+'/'+str(dfvalue[2])+'.dcm'
    shutil.copyfile(src, dst)
    
 
```

# dicom image to nifti image conversion

### Import libs

```python
import os
import nibabel as nib
import SimpleITK as sitk
```

### Conversion function from dicom 2 nii and save nii images

```python
output_nii_folder = '/kaggle/working/nii_img/'
if os.path.exists(output_nii_folder) == False:
    os.mkdir(output_nii_folder)

def dicom2nifti(main_dir, image_dir, save=True):
    "given a dicom directory, loads them into single file and can save it as .nii file"
    reader = sitk.ImageSeriesReader()
    reader.LoadPrivateTagsOn()
    file_dir = main_dir+image_dir
    filenamesDICOM = reader.GetGDCMSeriesFileNames(str(file_dir))
    reader.SetFileNames(filenamesDICOM)
    img = reader.Execute()
    img = sitk.Cast(img, sitk.sitkFloat32)
    
    if save:
        sitk.WriteImage(img, output_nii_folder+image_dir+'.nii.gz')
    else:
        return img
```

```python
main_input_dir = '/kaggle/working/train_imgs/'

for img_folder in os.listdir(main_input_dir):
    print(img_folder)
    dicom2nifti(main_input_dir, img_folder)
```
