```python
ids = [int(x.split('.')[0]) for x in test_image_list]
ids[:20]
```

```python
target = np.load('../input/cats-and-dogs-embedded-data/cats_and_dogs_1/target.npy')
```

```python
train_ENV2L = np.load('../input/cats-and-dogs-embedded-data/cats_and_dogs_5/cats_and_dogs_5/train_ENV2L.npy')
test_ENV2L = np.load('../input/cats-and-dogs-embedded-data/cats_and_dogs_5/cats_and_dogs_5/test_ENV2L.npy')
```

```python
train_ENV2L.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_ENV2L, target, test_size=0.1, random_state=42)
```

```python
%%time
svc = SVC(C= 0.5, probability=True)
svc.fit(X_train, y_train)
```

```python
%%time
val_preds = svc.predict_proba(X_val)[:,1]
```
