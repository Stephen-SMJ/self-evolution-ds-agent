```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
df = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
df.head(
```

```python
ds = df.to_numpy()
ds.shape
```

```python
y,X = np.split(ds, (1, ), axis = 1 )
y = y.squeeze()
X.shape, y.shape
```

```python
Xtrain, Xvalidation, Xtest = np.split( X, (35000, 38000), axis = 0 )
ytrain, yvalidation, ytest = np.split( y, (35000, 38000), axis = 0 )
Xtrain = Xtrain / 255   # 标准化处理
Xvalidation = Xvalidation / 255
Xtest = Xtest / 255
Xtrain.shape, Xvalidation.shape, Xtest.shape
```

```python
W = np.random.random( (784, 10) )
W0 = np.random.random( )
```

```python
t = 100
xue = 1e0
for i in range(5000):
    # 1、随机W和W0，进行线性计算，score
    score = Xtrain @ W + W0
    # 2、e**score, 计算归一化，概率P
    es = np.exp(score)
    P = (es.T / es.sum(axis=1) ).T
    # 2.1、计算onehot
    zeros = np.zeros( (10, 10) )
    zeros[np.arange(10), np.arange(10)] = 1
    onehot = zeros[ytrain]
    # 求 k 
    kw = ((P-onehot).T @ Xtrain / 42000).T
    kw0 = (P - onehot).mean(axis=0)
    # 更新W
    W = W - kw * xue
    W0 = W0 - kw0 * xue
    # 计算loss
    loss = - np.log( (P * onehot).sum(axis=1) ).mean()
#     if i % 100 == 99:
#         print(loss)
#         print(loss < t)
#         t = loss
```

```python
# SoftMax算法
# 1、随机W和W0，进行线性计算，score
score = Xvalidation @ W + W0
print(W.shape, W0.shape, score.shape)
# 2、e**score, 计算归一化，概率P
t = np.exp(score)
P = t / (t.sum(axis=1).reshape(-1,1)+1e-10)
# 3、argmax计算 Y' （YP）
YP = P.argmax(axis=1)
# print(YP.shape)
# # 计算正确率
# print((YP == yvalidation).sum()/3000)
```
